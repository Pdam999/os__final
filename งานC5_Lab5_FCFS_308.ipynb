2. การแสดงผลโดยกราฟเส้นและกราฟแท่ง เพื่อเปรียบเทียบค่า Average waiting time และค่า Average Turn Around time

# 1

```python
process_queue = []
waiting_time = 0
n = int(input('Enter the total no of processes: '))
turn_around_time = 0
for i in range(n):
    process_queue.append([])#append a list object to the list
    process_queue[i].append("P" + str(i)) # ProcessName
    process_queue[i].append("A" + str(i)) # Arrival time
    process_queue[i].append(int(input("P" + str(i) + ' burst time: '))) # Burst time
    if(i==0) :
      process_queue[i].append(0)               #Waiting time
    else:  
      process_queue[i].append(waiting_time)    #Waiting time 
    waiting_time += process_queue[i][2]
                       #Burst time           waiting time 
    turn_around_time = process_queue[i][2] + process_queue[i][3]
    process_queue[i].append(turn_around_time) #turn around time
    print("turn_around_time:",turn_around_time)
    print ('')
process_queue.sort(key = lambda process_queue:process_queue[1])
print ('ProcessName\tArrivalTime\tBurstTime\tWaitingTime\tTurnAroundTime')

total_waiting_time = 0 
turn_around_time = 0
total_turn_around_time = 0 
for i in range(n):
    #      ProcessName	               ArrivalTime	            BurstTime	                WaitingTime	                 TurnAroundTime
    print (process_queue[i][0],'\t\t',process_queue[i][1],'\t\t',process_queue[i][2],'\t\t',process_queue[i][3],'\t\t',process_queue[i][4])
    total_waiting_time += process_queue[i][3]
    turn_around_time = process_queue[i][2] + process_queue[i][3]
    total_turn_around_time += process_queue[i][4]
    print("turn_around_time:",turn_around_time)

print ( '-','\t\t','-','\t\t','total:','\t',total_waiting_time,'\t\t',total_turn_around_time)
print ( '-','\t\t','-','\t\t','Average:','\t',(total_waiting_time/n),'\t\t',(total_turn_around_time/n))
```

```python
import matplotlib.pyplot as plt
 
x  = [5, 10, 15, 20, 25] # จำนวน process - 5 ,10 ,15, 20, 25 
y1 = [0, 5, 15, 30, 50] # Average waiting time
y2 = [5, 15, 30, 50, 75] # Average turn around time
plt.plot(x, y1, label="line L")
plt.plot(x, y2, label="line H")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```
